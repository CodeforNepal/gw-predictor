### Loading necessary packages and preparing kernel.

```python
import pandas as pd
```

### Reading data
Shortned version of data file is used for github sharing. Complete data file can be provided when requested.

```python
data = pd.read_excel('data\DeepTubeWell_Data-SAMPLE.xlsx', sheet_name = 'Sheet1')
data.head()
```

### Extracting first row of every well.

```python
# data['well_id'].unique()
```

```python
screened_data = pd.DataFrame()
for id in data['well_id'].unique():
    screened_data = pd.concat([screened_data, data[data['well_id'] == id].iloc[:1, :] ])

screened_data.head()
```

### Filtering out rows containing no discharge values.

```python
screened_data = screened_data[screened_data['Q_lpm'].notna()]
screened_data.head()
```

### Creating a new dataframe with only necessary columns for discharge modeling

```python
qdata = screened_data.reset_index()[['well ','X_UTM','Y_UTM', 'Q_lpm']]
qdata
```

### Save new dataframe 

```python
# qdata.to_csv('data\cleaned-qdata.csv', index =False)
qdata.to_csv('data\cleaned-qdata-SAMPLE.csv', index =False)

# Saving disabled to overwrite existing file.
```
